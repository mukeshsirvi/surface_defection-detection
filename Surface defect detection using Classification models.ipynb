```python
def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
import os
path_to_data="./NEU Metal Surface Defects Data/train"
train_img_dir=[]
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        train_img_dir.append(entry.path)
```

```python
train_img_dir
```

```python
defects_file_names_dict = {}
for img_dir in train_img_dir[0]:
    defect_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(train_img_dir[0]):
        file_list.append(entry.path)
    defects_file_names_dict["crazing"] = file_list
defects_file_names_dict
```

```python
for img_dir in train_img_dir[1]:
    defect_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(train_img_dir[1]):
        file_list.append(entry.path)
    defects_file_names_dict["inclusion"] = file_list
defects_file_names_dict
```

```python
for img_dir in train_img_dir[2]:
    defect_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(train_img_dir[2]):
        file_list.append(entry.path)
    defects_file_names_dict["patches"] = file_list
defects_file_names_dict
```

```python
for img_dir in train_img_dir[3]:
    defect_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(train_img_dir[3]):
        file_list.append(entry.path)
    defects_file_names_dict["pitted-surface"] = file_list
defects_file_names_dict
```

```python
for img_dir in train_img_dir[4]:
    defect_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(train_img_dir[4]):
        file_list.append(entry.path)
    defects_file_names_dict["rolled-in_scale"] = file_list
defects_file_names_dict
```

```python
for img_dir in train_img_dir[5]:
    defect_name = img_dir.split('/')[-1]
    file_list = []
    for entry in os.scandir(train_img_dir[5]):
        file_list.append(entry.path)
    defects_file_names_dict["scratches"] = file_list
defects_file_names_dict
```

```python
class_dict = {}
count = 0
for defect_name in defects_file_names_dict.keys():
    class_dict[defect_name] = count
    count = count + 1
class_dict
```

```python
X, y = [], []
for defect_name, training_files in defects_file_names_dict.items():
    for training_image in training_files:
        img = cv2.imread(training_image)
        scalled_raw_img = cv2.resize(img, (32, 32))
        img_har = w2d(img,'db1',5)
        scalled_img_har = cv2.resize(img_har, (32, 32))
        combined_img = np.vstack((scalled_raw_img.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
        X.append(combined_img)
        y.append(class_dict[defect_name])   
```

```python
X = np.array(X).reshape(len(X),4096).astype(float)
X.shape
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

pipe = Pipeline([('scaler', StandardScaler()), ('svc', SVC(kernel = 'rbf', C = 10))])
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
print(classification_report(y_test, pipe.predict(X_test)))
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto',probability=True),
        'params' : {
            'svc__C': [1,10,100,1000],
            'svc__kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'randomforestclassifier__n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'logisticregression__C': [1,5,10]
        }
    }
}
```

```python
scores = []
best_estimators = {}
import pandas as pd
for algo, mp in model_params.items():
    pipe = make_pipeline(StandardScaler(), mp['model'])
    clf =  GridSearchCV(pipe, mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model': algo,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
best_clf = best_estimators['svm']
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, best_clf.predict(X_test))
cm
```
